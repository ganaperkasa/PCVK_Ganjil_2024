```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow  # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
urls = [
    "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
    "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
    "https://scontent.fmlg16-1.fna.fbcdn.net/v/t39.30808-6/311664281_184471900759664_5273547538134638512_n.jpg?_nc_cat=110&ccb=1-7&_nc_sid=6ee11a&_nc_eui2=AeFRpibSGz_yzJEigszyOlPB82OE9VZmDc7zY4T1VmYNzpOnRSal__kIXtvfSVVZJVof6ccKiIELP3XI_dX57Sxl&_nc_ohc=yQ55p18DW2AQ7kNvgF7iJCb&_nc_ht=scontent.fmlg16-1.fna&_nc_gid=ASmETikOX5OfxDmQ1e39hTm&oh=00_AYAfI3e8naTTbFc-PZvT1JNL5EPrC5cWi8FTQKeuc3NOiQ&oe=670AF81D"
]

# Read and display images
# Loop through each URL image, some images may be stored in a list
for url in urls:
    image = io.imread(url)  # read image
    image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)  # resize to half size
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  # convert color to RGB
    final_frame = cv.hconcat((image, image_2))  # concatenate images
    cv2_imshow(final_frame)  # show image
    print('\n')
```

3. Kode ini berfungsi untuk mengubah ukuran gambar (resize). Dalam kode tersebut, gambar image diubah skalanya (scale) dengan faktor fx=0.5 (lebar) dan fy=0.5 (tinggi). Artinya, gambar akan dikecilkan menjadi setengah ukuran aslinya. Jika kode ini tidak dilakukan, gambar akan tetap pada ukuran aslinya, yang mungkin terlalu besar atau tidak sesuai untuk keperluan uji coba selanjutnya.

4. Kode [255, 255, 255] merepresentasikan warna putih dalam format RGB (Red, Green, Blue). Kode ini digunakan untuk menggambar garis horizontal di tengah-tengah gambar dengan warna putih. Setiap piksel pada posisi (tinggi/2, y) akan diubah menjadi warna putih, membentuk garis horizontal.

5. Resolusi gambar ditentukan oleh jumlah piksel (unit terkecil dalam gambar). Semakin tinggi resolusi, semakin banyak jumlah piksel yang ada dalam gambar. Hal ini membuat gambar lebih tajam dan detail. Sebaliknya, jika resolusi rendah (jumlah piksel lebih sedikit), gambar akan terlihat buram atau tidak jelas karena detail yang bisa ditampilkan lebih sedikit. Jadi, resolusi yang tinggi umumnya lebih baik untuk tampilan visual, tetapi memerlukan ukuran file yang lebih besar dan lebih banyak daya komputasi untuk diproses.

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Draw horizontal line in the middle
for x in range(lebar):  # Iterate over the width
    image_3[x, int(lebar / 2)] = [255, 255, 255]

# Concatenate images and display
final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Draw diagonal line from top-left to bottom-right
for i in range(min(tinggi, lebar)):
    image_3[i, i] = [255, 255, 255]  # Diagonal line
    image_3[i, lebar - i - 1] = [255, 255, 255]

# Concatenate images and display
final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Draw shorter horizontal line in the middle (half the width)
start_y = int(lebar / 4)  # Starting point of the line
end_y = int(3 * lebar / 4)  # End point of the line

for y in range(start_y, end_y):
    image_3[int(tinggi / 2), y] = [255, 255, 255]

# Concatenate images and display
final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Draw a random white box (with specified top-left and bottom-right corners)
start_x_box = int(lebar / 3)
start_y_box = int(tinggi / 3)
end_x_box = start_x_box + 50  # Width of the box
end_y_box = start_y_box + 50  # Height of the box

for x in range(start_x_box, end_x_box):
    for y in range(start_y_box, end_y_box):
        image_3[y, x] = [255, 255, 255]  # White pixel

# Concatenate images and display
final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
urls = [
    "https://tangselife.com/wp-content/uploads/2024/03/2024-03-06-1.jpg"
]

# Loop through each URL image
for url in urls:
    image = io.imread(url)  # Read the image
    image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)  # Resize to half size
    image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Define the coordinates for the areas to cover outside the loop
# Adjust these values based on the actual position of NIK and name on the ID card
# Format: (start_x, start_y, end_x, end_y)
nik_area = (120, 50, 350, 75)  # Coordinates for the NIK area
name_area = (150, 100, 350, 115)  # Coordinates for the name area

# Cover the NIK area with a white rectangle
cv.rectangle(image, (nik_area[0], nik_area[1]), (nik_area[2], nik_area[3]), (255, 20, 25), -1)

# Cover the name area with a white rectangle
cv.rectangle(image, (name_area[0], name_area[1]), (name_area[2], name_area[3]), (65, 255, 55), -1)

# Show the modified image
cv2_imshow(image)  # Display the image with covered areas
print('\n')
```
